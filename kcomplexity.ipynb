```python
import itertools, ast
import warnings
from tqdm.auto import tqdm
warnings.filterwarnings("ignore")
from tqdm import TqdmWarning # or other specific tqdm warning classes
warnings.filterwarnings("ignore", category=TqdmWarning)

def get_results(alphabet, max_len):
    results = []
    errors = 0
    for L in range(max_len + 1):
        total = len(alphabet) ** L
        for p in tqdm(itertools.product(alphabet, repeat=L), total=total, desc=f"len={L}"):
            s = "".join(p)
            try:
                value = str(eval(s))
            except Exception as e:
                #print(f"Error evaluating '{s}': {e}")
                errors += 1
                continue
            results.append((s, value))
            
    return results, errors
          
alphabet = "0123+-*/()'"
max_len = 7
results, errors = get_results(alphabet, max_len)
print(f"Total results: {len(results)}, errors: {errors}")

def longest_result(results):
    longest = None
    for s, value in results:
        if longest is None or len(s) > len(longest):
            longest = s
    print("Longest result is:", longest, "with length", len(longest))

print("First 100 results:")
for result in results[:100]:
    print(result[0], "=>", result[1])

longest_result(results)
```

```python
def search_results(results, target):
    print("Expressions that evaluate to", target+":")
    count = 0
    shortest = None
    for s, value in results:
        if value == target:
            count += 1
            print(s, "=>", value)
            if shortest is None or len(s) < len(shortest):
                shortest = s
    print("Total found:", count)
    if shortest is not None:
        print(f'Shortest expression is "{shortest}" with length', len(shortest))
               
target = "1000000"
search_results(results, target)
```
